```python
import plotly as py
import json
import urllib.request
```

```python
country_code= json.load(open('country_code.json'))
country_num=  json.load(open('country_data_number.json'))
country_ratings= json.load(open('country_data_ratings.json'))
countries_del_list=['CSHH', 'CSXX', 'DDDE', 'SUHH', 'YUCS']
for country in countries_del_list:
    del country_code[country]
    del country_num[country]
    del country_ratings[country]
    
'''
Transfer 2 digit country code to 3 digit country code
'''
code_mapping={}
with urllib.request.urlopen("http://country.io/iso3.json") as url:
    code_mapping = json.loads(url.read().decode())

codes=[]
nums=[]
ratings=[]
names=[]
for key in country_code.keys():
    codes.append(code_mapping.get(key))
    names.append(country_code.get(key))
    nums.append(country_num.get(key))
    ratings.append(country_ratings.get(key))


```

```python
map_data=[dict(
    type='choropleth',
    locations=codes,
    z=nums,
    text=names,
    colorscale=[[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
    autocolorscale= True,
    marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Movie Number'),
      )
]

layout= dict(
    title= "Movie Number Distribution",
    geo=dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

import plotly as py
py.offline.init_notebook_mode(connected=True)
fig = dict(data=map_data, layout=layout)
py.offline.iplot(fig, validate=False, filename='number_distribution.html' )
# or bigger picture(recommended)
py.offline.plot(fig, validate=False, filename='number_distribution.html' )
```

```python
codes_temp=[]
names_temp=[]
ratings_temp=[]
for code, name, rating in zip(codes, names, ratings):
    if rating==0:
        continue
    codes_temp.append(code)
    names_temp.append(name)
    ratings_temp.append(rating)

        

map_data=[dict(
    type='choropleth',
    locations=codes_temp,
    z=ratings_temp,
    text=names_temp,
    colorscale=[[0,"#ffffff"], [1,"#004d39"]],
    
    autocolorscale= False,
    marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Movie Rating'),
      )
]

layout= dict(
    title= "Movie Rating Distribution",
    geo=dict(
        showframe = True,
        showcoastlines = True,
        projection = dict(
            type = 'Mercator'
        )
    )
)

import plotly as py
py.offline.init_notebook_mode(connected=True)
fig = dict(data=map_data, layout=layout)
py.offline.iplot(fig, validate=False, filename='rating_distribution.html' )
# or bigger picture
# py.offline.plot(fig, validate=False, filename='rating_distribution.html' )
```
